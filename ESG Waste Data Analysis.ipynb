```python
df2 = pd.get_dummies(df, columns=['Company Name'], drop_first=True)
```

```python
df2.head()
```

```python
import statsmodels.api as sm
```

```python
x = df2[['Total Waste Generated (MT)', 'Plastic Waste (MT)', 'Ewaste (MT)','Bio Medical Waste (MT)','Construction and Demolition Waste (MT)','Battery Waste (MT)','Radioactive Waste (MT)','Other Hazardous Waste (MT)','Other NonHazardous Waste Generated (MT)']]
y= df2['Total Scope 1 and Scope 2 Emissions (MTCO₂e)']
```

```python
# Add a constant to the independent variables matrix for the intercept
x = sm.add_constant(x)
```

```python
# fit the regression model:-
model= sm.OLS(y,x).fit()
```

```python
summary= model.summary()
```

```python
print(summary)
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data["Variable"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]

# Print the VIF dataframe
print(vif_data)
```
